```python
class matrix:
    def __init__(self, l_o_l):
        self.__l_o_l = l_o_l
        self.shape_check()
        
    def shape_check(self):
        for i in range(len(self.__l_o_l) - 1):
            if len(self.__l_o_l[i]) != len(self.__l_o_l[i + 1]):
                raise ValueError("Matrix is an invalid shape.")
            
    def __len__(self): #returns matrix length/# of rows
        return len(self.__l_o_l)
            
    def __repr__(self):
        return repr(self.__l_o_l)
            
    def __getitem__(self, key):
        if isinstance(key, tuple):
            row, col = key
            return self.__l_o_l[row][col]
        return self.__l_o_l[key]
    
    def assign(self, other):
        if isinstance(other, matrix):
            data = other.__l_o_l
        else:
            data = other
        if len(data) != len(self.__l_o_l) or len(data[0]) != len(self.__l_o_l[0]):
            raise ValueError("Matrix sizes do not match.")

        for i in range(len(self.__l_o_l)):
            for j in range(len(self.__l_o_l[0])):
                self.__l_o_l[i][j] = data[i][j]
    
    def shape(self): #returns matrix dimensions as a tuple
        return (len(self.__l_o_l), len(self.__l_o_l[0]))
    
    def transpose(self):
        M_T = []
        for i in range(len(self.__l_o_l[0])): #loops over row length
            row = []
            for j in range(len(self.__l_o_l)): #then column length
                row.append(self.__l_o_l[j][i]) #to rearrange and form the transposed matrix
            M_T.append(row)
        return matrix(M_T)
    
    def row(self, n):
        return matrix([self.__l_o_l[n]]) #returns row vector

    def col(self, n): 
        col = []
        for i in range(len(self.__l_o_l)):
            col.append([self.__l_o_l[i][n]]) #gets column values and returns column vector
        return matrix(col)

    def to_list(self):
        return self.__l_o_l #makes matrix type back to list
        
    def block(self, n_0, n_1, m_0, m_1):
        block = []
        for i in range(n_1 - n_0 + 1): #calculates size of block, then loops over it to return values inside
            row = []
            for j in range(m_1 - m_0 + 1):
                row.append(self.__l_o_l[n_0 + i][m_0 + j])
            block.append(row)
        return matrix(block)

M_1 = matrix([[1, 2, 3], [4, 5, 6]]) #tests exercise 2 functions
print(M_1.shape())
print(M_1.transpose())
print(M_1.row(1))
print(M_1.col(1))
print(type(M_1.to_list()))
print(M_1.block(0, 0, 1, 2))
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    result = []
    for i in range(n): #loops over matrix size and places each value as the constant inputted
        row = []
        for j in range(m):
            row.append(c)
        result.append(row)
    return matrix(result)

def zeros(n, m): #calls on constant function to return a zero matrix
    return constant(n, m, 0)

def ones(n, m): #calls on constant function to return a ones matrix
    return constant(n, m, 1)

def eye(n): #calls on zero matrix then replaces diagonal elements with 1 to return an identity matrix
    result = zeros(n, n)
    for i in range(n):
        result[i][i] = 1
    return matrix(result)

print(constant(2, 2, 44)) #tests exercise 3 functions
print(zeros(2, 2))
print(ones(2, 2))
print(eye(3))
```
